## 🧠 Challenge: Find the Top Month and Product

- Which month had the highest total sales?
- Which product sold the most overall?

(Hint: use `.sum()` and `.idxmax()`)

Try writing your own code below.


````python
import pandas as pd  # pd is a common shortcut name.

# Create a dictionary of lists
data = {
    'Month': ['January', 'February', 'March'],
    'Product A': [120, 134, 145],
    'Product B': [90, 100, 110],
    'Product C': [140, 150, 160]
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Show the table
print("Original DataFrame:")
print(df)

# Print just the 'Product B' column
print("\n'Product B' column:")
print(df['Product B'])

# Print the second row (index 1)
print("\nSecond row:")
print(df.iloc[1])

# Add a column for total sales across all products
df['Total Sales'] = df[['Product A', 'Product B', 'Product C']].sum(axis=1)

# Add a column for average sales
df['Average Sales'] = df[['Product A', 'Product B', 'Product C']].mean(axis=1)

# Show the updated table
print("\nUpdated DataFrame with totals and averages:")
print(df)

# Challenge: Find the top month and product

# Month with the highest total sales
top_month = df.loc[df['Total Sales'].idxmax(), 'Month']
print(f"\nMonth with the highest total sales: {top_month}")

# Product that sold the most overall
product_totals = df[['Product A', 'Product B', 'Product C']].sum()
top_product = product_totals.idxmax()
print(f"Product with the highest total sales: {top_product}")
````
